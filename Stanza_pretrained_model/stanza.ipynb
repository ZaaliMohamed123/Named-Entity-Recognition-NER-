```python
import stanza
```

```python
#Initialiser le pipeline (Exemple en anglais)
stanza.download('en')  # Télécharge le modèle anglais 
nlp = stanza.Pipeline('en')  # Charger le pipeline NLP anglais
```

```python
text = "Barack Obama was born in Hawaii and served as the 44th President of the United States."
```

```python
doc = nlp(text)
```

```python
print("Named Entities:")
for entity in doc.entities:
    print(f"Text: {entity.text}, Type: {entity.type}, Start: {entity.start_char}, End: {entity.end_char}")
```

```python
def stanza_ner(sentence):
    
    doc = nlp(sentence)
    entities = {entity.text:entity.type for entity in doc.entities}


    return entities
```

```python
stanza_ner("Tesla Inc is going to acquire Twitter Inc for $45 billion")
```
